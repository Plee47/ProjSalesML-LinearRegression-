# Sales Quantity Prediction Based on Marketing Campaign Value (Machine Learning)

Objective: Make a projection of inventory needs based on the marketing investment value (Linear Regression)

```python
#importing librarys

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model as lm
```

```python
#reading a simple database

data = pd.read_excel('InvestimentxSales.xlsx')
```

```python
#displaying the base graphically

plt.plot(data['Investment value'], data['Sales quantity'])
plt.ylabel('Sales quantity')
plt.xlabel('Investment value')
plt.grid(True)
plt.show()
```

```python
#applying ml to the base

reg = lm.LinearRegression()

reg.fit(data['Investment value'].values.reshape(-1, 1), data['Sales quantity'])
```

```python
#drawing a line between the points using the parameters (better line)

plt.scatter(data['Investment value'], data['Sales quantity'])
x = np.array(data['Investment value'])
y = reg.intercept_ + x * reg.coef_
plt.plot(x, y, 'r')
plt.show()
```

```python
#example using 122000 on investiment

reg.predict([[122000]])
```

```python
#checking accuracy with a graph

plt.scatter(data['Investment value'], data['Sales quantity'])
plt.scatter(122000, reg.predict([[122000]])[0], color='k')
x = np.array(data['Investment value'])
y = reg.intercept_ + x * reg.coef_
plt.plot(x, y, 'r')
plt.show()
```
